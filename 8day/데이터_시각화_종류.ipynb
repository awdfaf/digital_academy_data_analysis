```python
# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np

# seaborn 내장 데이터 읽어오기
# 데이터 셋 : anscombe
# 변수이름 : ans
ans = sns.load_dataset("anscombe")
ans.info()
```

```python
# dataset 컬럼의 값을 기준으로 그룹화
# 변수명 : ans_group
ans_group = ans.groupby(["dataset"])
ans_group.head()
```

```python
# 기초통계확인
ans_group.describe()
```

```python
# 시각화 라이브러리
import matplotlib.pylab as plt

# ans 데이터 원본을 이용해 
# dataset 컬럼의 데이터 중 I 값만 추출
data1 = ans[ans["dataset"] == "I"]
data1
```

```python
# dataset = I에 대한 그래프 그리기
plt.plot(data1["x"],data1["y"], "o")
```

```python
# dataset 다른 데이터에 대해서도 필터링 후 그래프 그리기
data2 = ans[ans["dataset"] == "II"]
data3 = ans[ans["dataset"] == "III"]
data4 = ans[ans["dataset"] == "IV"]
```

```python
# 그래프 그리기 위한 객체 받아오기
fig = plt.figure()

# 4개 각각의 그래프를 그리기 위한 공간만들기
# subplot
# 첫번째 값 : 행
# 두번째 값 : 열
# 세번째 값 : 행렬 중 어느 위치에 들어갈지 순서 지정
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# subplot에 제목넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")



# 데이터 넣어서 그래프 그리기 : 점으로 표시
ax1.plot(data1["x"],data1["y"], "o",c="r")
ax2.plot(data2["x"],data2["y"], "o",c="g")
ax3.plot(data3["x"],data3["y"], "o",c="b")
ax4.plot(data4["x"],data4["y"], "o",c="y")

# 전체 그래프 제목
fig.suptitle("Anscombe Data")

# 그래프 간 간격 조절
# tight_layout()
fig.tight_layout()

# 우상향 데이터 /우하향 데이터 
```

```python
# seaborn으로 시각화
# seaborn에서 제공해주는 데이터셋중 주로 사용하는 데이터
# 붓꽃 데이터
iris = sns.load_dataset("iris")

# 타이타닉 호
titanic = sns.load_dataset("titanic")

# 팁 
tips = sns.load_dataset("tips")

# 여객운송
flights = sns.load_dataset("flights")
```

```python
# stripplot 시각화
# 기본 팔레트 배경 설정
sns.set_palette("pastel")

# stripplot 모양 지정
sns.stripplot(x="day",y="total_bill",data=tips)

plt.title("팁 데이터")
plt.xlabel("요일")
plt.ylabel("요금")

plt.show()
```

```python
# 막대그래프
# 평균치를 빠르게 집계해서 시각화 가능
# 막대그래프 위 선 : 오차막대(error bar)
# 오차막대는 막대그래프 위에서 끝막대를 기준으로 반반씩
# 오차막대의 길이가 작을수록 좋다
sns.barplot(x="sex",y="tip",data=tips)
```

```python
# 막대그래프 오차막대 숨기기 : estimator=len
sns.barplot(x="sex",y="tip", estimator=len,data=tips)
```

```python
# 3개 컬럼의 데이터 사용
# 성별/요일별 tip 데이터 시각화
sns.barplot(x="sex",y="tip",hue="day",data=tips)
```

```python
# 박스플롯 시각화
sns.boxplot(x="sex",y="tip",data=tips)
```

```python
# 박스플롯 눕히기

# 숫자 데이터 박스플롯
# orient = "v" : 수직 (디폴트)
# orient = "h" : 수평
sns.boxplot(data=tips,orient="h")
```

```python
# 커널밀도그래프
# 히스토그램보다 부드러운 형태의 분포 곡선으로 시각화
sns.kdeplot(tips.tip)
```

```python
# 커널밀도그래프와 히스토그램을 하나의 그래프로 표현하는 함수
# 선 벗어나 있는 데이터는 이상치일 수 있다
sns.distplot(tips.tip)
```

```python
sns.histplot(tips.tip)
```

```python
# countplot : 범주형 데이터 표시할 때 주로 사용
# data 속성에는 데이터프레임만 사용 가능
# 특정 컬럼의 범주 데이터 별 빈도 확인 가능(값이 얼마나 들어 있는지 count와 유사)
sns.countplot(x="day",data=tips)
```

```python
# 여러 그래프를 하나로 그리기
# kind : 그래프 종류 지정
# scatter : 산점도와 막대 그래프를 동시 시각화
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
```

```python
# 여러 그래프를 하나로 그리기
# kind : 그래프 종류 지정
# kde : 커널밀도와 선 그래프를 동시 시각화
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
# 다차원 복합 데이터 시각화
# swarm : 데이터를 나타내는 점이 겹치지 않도록 표현
# violin : swarm의 점 전체 모양을 바이올린 모양으로 표현
# catplot : 색상과 행 등을 동시에 사용하여 3개 이상의 범주값의 분포 변화를 시각화
sns.catplot(x="day",y="total_bill",hue="sex",kind="swarm",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",kind="violin",data=tips)
```

```python
# 바이올린플롯
# 두개의 컬럼 데이터만 사용가능
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
# 다차원 시각화 : 여러 컬럼 데이터 시각화
# 크로스 체크 가능
sns.pairplot(tips)
```

```python
# 다차원 상관도 
sns.pairplot(tips,hue="sex")
```
