```python
# 4_4_seoul_starbucks_list 파일 읽기
# 변수명 : seoul_starbucks

# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np

file_path = "./data/4_4_seoul_starbucks_list.xlsx"
seoul_starbucks = pd.read_excel(file_path)
seoul_starbucks
```

```python
# 정보확인
seoul_starbucks.info()
```

```python
# 지도시각화 라이브러리 불러들이기
# 프롬폼트에서 설치
import folium
```

```python
# 데이터형태 중 json형태의 데이터 처리를 위한 라이브러리
# json 데이터 형태는 일반적으로 딕셔너리 형태
import json
```

```python
# 지도 그리기
starbucks_map = folium.Map(
    # 최초의 보여줄 지도 위치 지정(지도 중심점 좌표)
    # [위도, 경도] 지정
    location = [37.573050, 126.979189],
    # 지도 스타일 지정 ; tiles
    # openstreetmap : 도시형 건물 등 일반적으로 사용되는 스타일 
    # Stamen Terrain : 산림 위주로 보여주는 스타일
    # cartodbposition : 하천이나 도로(길) 위주로 보여주는 스타일
    tiles = "openstreetmap",
    # 최초의 보여질 지도 zoom 사이즈 지정 
    zoom_start = 50
)
starbucks_map
```

```python
# 스타벅스 매장 위치 표시
# 지도 시각화 시 여러 좌표의 위치를 표시할 때 반복문 사용
for idx in seoul_starbucks.index:
    # 위/경도 데이터 변수처리
    lat = seoul_starbucks.loc[idx,"위도"]
    lng = seoul_starbucks.loc[idx,"경도"]
    
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        # 마킹 위치 지정
        location=[lat,lng],
        fill_color="green",
        fill_opacity=1,
        color="yellow",
        weight=1,
        radius=3

        # 사용할 지도형태 지정 : 위에서 이미 지정했음
    ).add_to(starbucks_map)

starbucks_map
```

```python
# 매장 타입 별 위치 표시


# 지도 그리기
starbucks_map2 = folium.Map(
    # 최초의 보여줄 지도 위치 지정(지도 중심점 좌표)
    # [위도, 경도] 지정
    location = [37.573050, 126.979189],
    # 지도 스타일 지정 ; tiles
    # openstreetmap : 도시형 건물 등 일반적으로 사용되는 스타일 
    # Stamen Terrain : 산림 위주로 보여주는 스타일
    # cartodbposition : 하천이나 도로(길) 위주로 보여주는 스타일
    tiles = "openstreetmap",
    # 최초의 보여질 지도 zoom 사이즈 지정 
    zoom_start = 11
)

for idx in seoul_starbucks.index:
    # 위/경도 데이터 변수처리
    lat = seoul_starbucks.loc[idx,"위도"]
    lng = seoul_starbucks.loc[idx,"경도"]
    store_type = seoul_starbucks.loc[idx,'매장타입']
    
    # 매장타입별 표시할 색상 지정
    fillColor = ""
    size = 1
    if store_type == "general":
        fillColor = "gray"
        size = 3
    elif store_type == "reserve":
        fillColor = "blue"
        size = 5
    elif store_type == "generalDT":
        fillColor = "red"
        size = 5
        
    # 표시할 마커 스타일 지정
    folium.CircleMarker(
        # 마킹 위치 지정
        location=[lat,lng],
        fill_color=fillColor,
        fill=True,
        
        fill_opacity=1, # 불투명도
        color=fillColor,
        weight=1,
        radius=3

        # 사용할 지도형태 지정 : 위에서 이미 지정했음
    ).add_to(starbucks_map2)
```
