```python
# 데이터 읽어 들이기
# 변수명 : chatbot_data
# 수집 파일명 : total.xlsx
import pandas as pd
chatbot_data = pd.read_excel("./DataSet/total.xlsx")
chatbot_data
```

```python
# rule의 데이터를 list형태로 변환
# key : value의 딕셔너리 형태로 만들기
# key = index값 사용(0번,,,)
# value = 데이터프레임의 rule 데이터
# 딕셔너리 변수 : chat_dic = {}
chat_dic = {}
for i, rule in enumerate(chatbot_data["rule"]):
    a = []
    a.append(rule)
    chat_dic[i] = a
    
```

```python
chat_dic = {}
row = 0
for rule in chatbot_data["rule"] :
    chat_dic[row] = rule.split("|")
    row += 1
print(chat_dic)
```

```python
### [질문]이 데이터프레임의 request에 포함되어 있는지 확인하는 기능 구현
# - 있다면 : 데이터프레임에서 해당 index의 response의 값으로 응답 처리
# - 없다면 : "무슨 말인지 잘 모르겠네요~" 응답 처리
# ** 챗봇은 질문자와 반복해서 주고 받아야 하기 때문에..
#    - 질문을 처리하는 기능, 즉 함수로 만들어져 있어야 합니다.

def chat(request) :
    # request = "책 추천해줘"
    for k, v in chat_dic.items() :  # rule 기준으로 돎
        # print(k, v)
        ### rule에 포함되어 있는지 확인하기 위한 변수 지정
        chat_flag = False

        ### 질문 내용이 rule에 포함되어 있는지 확인하기
        for word in v :  # v값은 리스트, rule에 있는 데이터 개수만큼 돎
            if word in request :  # request는 문자열
                chat_flag = True
            else :
                chat_flag = False
                break

        ### 질문에 rule이 있다면...
        if chat_flag :
            return chatbot_data["response"][k]

    ### 질문에 대한 rule이 없다면
    return "무슨 말인지 잘 모르겠네요!"
```

```python
#챗봇 인터페이스 만들기
# 질문자와 응답 챗봇간 인터페이스
# 질문자의 질문을 함수에 전달
# 함수가 리턴한 응답을 질문자에게 보여주기
# 인터페이스 종료까지

# 무한 반복시키기
while True:
    try:
        # input() : 키보드로부터 입력을 받게끔하는 인터페이스 함수
        # 키보드로 입력한 모든 값(숫자 포함)은 문자열로 반환
        req = input("나 : ")
        
        # 질문자가 입력한 값 처리하기
        # 입력값이 exit이면 반복 종료
        if req == "exit":
            break
        else:
            # 질문을 함수에 보내서 응답 메세지 받아오기
            print("ChatBot : ", chat(req))
    except:
        print("알 수 없는 오류 발생")
        break
        
```
