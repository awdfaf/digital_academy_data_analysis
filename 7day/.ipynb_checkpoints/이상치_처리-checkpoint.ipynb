```python
# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
# 파일 불러오기
# encoding : 수집데이터가 utf-8 포멧이 아닌경우 오류발생 / 한글 데이터가 주로 오류 발생 -> euc-kr 설정
file_path = "./data/bicycle_out.csv"
df = pd.read_csv(file_path,encoding="euc-kr")
df
```

```python
df.info()
```

```python
# 결측치 처리하기
# 결측치가 있는 모든 행 삭제
# 변수명 : df_drop_allrow
df_drop_allrow = df.dropna(axis=0)
#df_drop_allrow.info()
df_drop_allrow
```

```python
# 인데스 번호 재배열
df_drop_allrow = df_drop_allrow.reset_index(drop=True)
df_drop_allrow
```

```python
# 이상치 데이터 확인
# 박스 플롯 시각화
plt.boxplot(df_drop_allrow.나이)
plt.show()
```

```python
# 기초통계 확인
df_drop_allrow.나이.describe()
```

```python
# 나이가 100살보다 큰 값에 대한 대이터 조회
df_drop_allrow[df_drop_allrow.나이 > 100]
```

```python
df_drop_allrow[(df_drop_allrow["나이"] > 100) | (df_drop_allrow["나이"] <= 10)]
```

```python
# 최대/최소 이상치 계산하여 제거
# q1(25%), q3(75%) 시점 데이터 값 확인
df_drop_allrow.나이.describe()
```

```python
# 넘파이 함수를 이용해서 특정 퍼센트에 해당 시점의 데이터 확인
# q1 : 최소이상치 한계값  q3 : 최대이상치 한계값
q1, q3 = np.percentile(df_drop_allrow.나이, [25,75])
print("q1=",q1," / q3=",q3)
```

```python
# IQR 계산
iqr = q3 - q1
print(iqr)
```

```python
# 최대/최소 이상치 한계 계산공식
# 최대 이상치 한계 = q3 + (iqr * 1.5)
# 최소 이상치 한계 = q1 - (iqr * 1.5)

# 최대 이상치 한계
upper_bound = q3 + (iqr *1.5)
print(upper_bound)
```

```python
# 최소 이상치 한계
lower_bound = q1 - (iqr *1.5)
print(lower_bound)
```

```python
# 최대/최소 이상치 한계를 벗어나는 데이터 조회
len(df_drop_allrow[(df_drop_allrow["나이"] > upper_bound) | (df_drop_allrow["나이"] < lower_bound)])
```

```python
condition = (df_drop_allrow["나이"] > upper_bound) | (df_drop_allrow["나이"] < lower_bound)
df_drop_allrow[condition]
```

```python
# 함수로 만들기
# 0. 함수이름 : outliers_iqr()
# 1, q1, q3 값 구하기
# 2. iqr 값 구하기
# 3. 최대 이상치 한계값 
# 4. 최소 이상치 한계값
# 5. 최대/최소 이상치 한계 벗어나는 데이터 리턴

def outliers_iqr(df,col):
    
    q1, q3 = np.percentile(df[col], [25,75])
    print("q1=",q1," / q3=",q3)
    iqr = q3 - q1
    print(iqr)
    upper_bound = q3 + (iqr *1.5)
    print(upper_bound)
    lower_bound = q1 - (iqr *1.5)
    print(lower_bound)

    return df_drop_allrow[(df_drop_allrow[col] > upper_bound) | (df_drop_allrow[col] < lower_bound)]
```

```python
# 결측치 제거 데이터프레임 변수 저장
df1 = df_drop_allrow
outliers_iqr(df1,"나이")
```

```python
## 필터링 조건에 만족하는 인덱스 값만 추출하는 함수 사용
# where() : 넘파이 함수, 배열로 리턴해 줍니다.
# 넘파이 배열과 파이썬 리스트 차이점
# 파이썬의 리스트 : 여러 타입의 데이터 값을 넣을 수 있다.
# 넘파이의 배열 : 한 가지 타입의 데이터 값만 넣을 수 있다.
# 위 차이점 외에 리스트와 사용법 및 기호가 모두 동일하다.

np.where((df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"] > upper_bound))
```

```python
# 범위를 벗어나는 데이터들의 인덱스 구하는 함수


def getOutliers_lqr(data):
    
    # q1, q3 구하기
    q1, q3 = np.percentile(data, [25,75])
    print("q1 = ",q1, "\nq3 = ", q3)
    
    # IQR 계산
    iqr = q3 - q1
    print("iqr = ",iqr)
    
    # 최대 이상치 한계 값
    upper_bound = q3 + (iqr * 1.5)
    print("upper_bound = ",upper_bound)
    
    # 최소 이상치 한계 값
    lower_bound = q1 - (iqr * 1.5)
    print("lower_bound = ",lower_bound)
    
#     # 최대 / 최소 이상치 한계를 벗어나는 데이터 조회
#     outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
#     print("outliers 의 길이 = ",len(outliers))
    
    return(np.where((data < lower_bound) | (data > upper_bound)))
```

```python
getOutliers_lqr(df_drop_allrow["나이"])

# 튜플 형태의 결과가 나온다.
```

```python
df_drop_allrow.loc[getOutliers_lqr(df_drop_allrow["나이"])]
```

```python
age_outlier_index = getOutliers_lqr(df_drop_allrow["나이"])
age_outlier_index
```

```python
# age_outlier_index 를 얻었다.
# 이 인덱스 번호를 삭제해야 한다.
    # 정상 데이터만 따로 담아낸다.
    # 원본의 데이터 인덱스 번호를 알아야 한다.
        # index 함수를 사용

df_drop_allrow.index
```

```python
# 깨끗한 데이터프레임 변수 : age_non_outlier_index
# 전체 데이터프레임에서 이상치 10개를 제외하고
# 깨끗한 데이터프레임에 담기

age_non_outlier_index = pd.DataFrame()

for i in df_drop_allrow.index:
    #print(i) 잘 나오는지 확인
    
    # 아닌 조건 / loc에 대괄호 두개
    if i not in age_outlier_index[0]:
        age_non_outlier_index = pd.concat([age_non_outlier_index, df_drop_allrow.loc[[i]]],
                                           ignore_index=True)
```
