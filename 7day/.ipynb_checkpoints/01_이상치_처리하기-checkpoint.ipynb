# 최대/최소 이상치 계산하여 제거하기

```python
import numpy as np
```

```python
# q1(25%), q3(75%) 시점의 데이터 값 확인하기

df_drop_allrow["나이"].describe()
```

```python
# 넘파이 함수를 이용하여 특정 퍼센트에 해당하는 시점의 데이터 확인하기
# q1 : 최소 이상치 한계값, 25%
# q3 : 최대 이상치 한계값, 75%

q1, q3 = np.percentile(df_drop_allrow["나이"], [25,75])
print("q1 = ",q1, "\nq3 = ", q3)
```

```python
# IQR 계산
iqr = q3 - q1
print(iqr)
```

```python
# 최대/ 최소 이상치 한계 계산공식
# 최대 이상치 한계 = q3 + (iqr * 1.5)
# 최소 이상치 한계 = q1 - (iqr * 1.5)
```

```python
# 최대 이상치 한계 값

upper_bound = q3 + (iqr * 1.5)
print(upper_bound)
```

```python
# 최소 이상치 한계 값

lower_bound = q1 - (iqr * 1.5)
print(lower_bound)
```

```python
# 최대 / 최소 이상치 한계를 벗어나는 데이터 조회

df_drop_exclude = df_drop_allrow[(df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"] > upper_bound)]
print(len(df_drop_exclude))
```

```python
### 함수로 만들기
# 함수에 들어갈 내용
# 1. q1, q3 값 구하기
# 2. iqr 값 구하기
# 3. 최대/ 최소 이상치 한계 값 구하기
# 4. 최대/ 최소 이상치 한계를 벗어나는 데이터 리턴하기

def outliers_iqr(dataframe, col):
    
    # q1, q3 구하기
    q1, q3 = np.percentile(dataframe[col], [25,75])
    print("q1 = ",q1, "\nq3 = ", q3)
    
    # IQR 계산
    iqr = q3 - q1
    print("iqr = ",iqr)
    
    # 최대 이상치 한계 값
    upper_bound = q3 + (iqr * 1.5)
    print("upper_bound = ",upper_bound)
    
    # 최소 이상치 한계 값
    lower_bound = q1 - (iqr * 1.5)
    print("lower_bound = ",lower_bound)
    
    # 최대 / 최소 이상치 한계를 벗어나는 데이터 조회
    outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
    print("outliers 의 길이 = ",len(outliers))
    
    return(outliers)
```

```python
outliers_iqr(df_drop_allrow, "나이")
```

```python
## 필터링 조건에 만족하는 인덱스 값만 추출하는 함수 사용
# where() : 넘파이 함수, 배열로 리턴해 줍니다.
# 넘파이 배열과 파이썬 리스트 차이점
# 파이썬의 리스트 : 여러 타입의 데이터 값을 넣을 수 있다.
# 넘파이의 배열 : 한 가지 타입의 데이터 값만 넣을 수 있다.
# 위 차이점 외에 리스트와 사용법 및 기호가 모두 동일하다.

np.where((df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"] > upper_bound))
```

```python
# 범위를 벗어나는 데이터들의 인덱스 구하는 함수


def getOutliers_lqr(data):
    
    # q1, q3 구하기
    q1, q3 = np.percentile(data, [25,75])
    print("q1 = ",q1, "\nq3 = ", q3)
    
    # IQR 계산
    iqr = q3 - q1
    print("iqr = ",iqr)
    
    # 최대 이상치 한계 값
    upper_bound = q3 + (iqr * 1.5)
    print("upper_bound = ",upper_bound)
    
    # 최소 이상치 한계 값
    lower_bound = q1 - (iqr * 1.5)
    print("lower_bound = ",lower_bound)
    
#     # 최대 / 최소 이상치 한계를 벗어나는 데이터 조회
#     outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
#     print("outliers 의 길이 = ",len(outliers))
    
    return(np.where((data < lower_bound) | (data > upper_bound)))
```

```python
getOutliers_lqr(df_drop_allrow["나이"])

# 튜플 형태의 결과가 나온다.
```
