```python
# 성별 수학점수의 평균 조회 : 데이터프레임 형태로 조회
df.groupby("sex")[["math"]].mean()
```

```python
# 성별 그룹화하기
# 변수 sex_group
sex_group = df.groupby("sex")
sex_group.head()
```

```python
# 그룹의 내부구조 확인 / 리스트에 있는 값은 인덱스 값
sex_group.groups
```

```python
# 위 그룹의 형태를 이용해 
# m에 있는 모든 인덱스값을 추출
for i in range(len(sex_group.groups["m"])):
    print(sex_group.groups["m"][i])
```

```python
for i in sex_group.groups["m"]:
    print(i)
```

```python
# 데이터 재구조화 / 특정 컬럼의 데이터를 특정구간으로 나누기
# 동일(유사)한 길이(범위)로 나눈다
df1 = pd.cut(df["math"],3)
df1
```

```python
# 구간별 소속 알려주기 / 라벨링 / 인덱스 번호처럼 0부터시작하는 번호 부여
df1 = pd.cut(df["math"],3,labels=False)
df1
```

```python
# 수학점수를 구간별로 그룹화
df2 = df["math"].groupby(df1)

df2.head(1) # 각 그룹의 첫번째 뽑기
```

```python
# 그룹화한 데이터 기초통계
# agg() 함수사용
df2.agg(["count","mean","std","min","max"])
```

```python
# 유사한 값들(동일하지 않은 범위)끼리 구간 나누기
# 구간 많아지면, 해당 구간에 없는 값이 생길 수 있다
# 해당 구간에 존재하는 값이 없을 경우 조회 안됨
pd.qcut(df["math"],5)
```

```python
pd.qcut(df["math"],5,labels=False)
```

```python
# 구간 값 지정
pd.qcut(df["math"],5,labels=np.arange(1,6,1))
```

```python
# 원-핫 인코딩 : 범주를 컬럼으로 데이터화 하고자 할 때 사용
# 0과 1의 값을 사용
df_onehot = pd.get_dummies(df)
df_onehot
```

```python
# 특정 컬럼에 대한 원-핫 인코딩
df_onehot2 = pd.get_dummies(df["sex"])
df_onehot2
```

```python
# 성별로 원핫인코딩한 변수를 이용해서 
# 성별로 몇명씩 있는지 확인 / count는 행 개수 세는거임
df_onehot2.sum()
```
