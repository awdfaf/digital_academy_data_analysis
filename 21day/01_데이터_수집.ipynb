```python
# csv파일 사용
# bicycle.csv 파일 읽어들이기
# 데이터프레임 변수 : df1
# 상위 5개 출력
import pandas as pd

file_path1 = "./data/bicycle.csv"
df1 = pd.read_csv(file_path1,encoding="euc-kr")
df1.head(5)
```

```python
# 엑셀파일 불러들이기

file_path2 = "./data/bicycle.xlsx"
df2 = pd.read_excel(file_path2)
df2.head(5)
```

```python
# read.json 파일 읽어들이기
file_path3 = "./data/read.json"
df3 = pd.read_json(file_path3)
df3.head(5)
```

```python
# 데이터프레임을 csv로 저장하기
# data폴더에 저장
# 저장파일명 : bicycle_save.csv
# 저장할 데이터프레임 : df1
df1.to_csv("./data/bicycle_save.csv", index = False)
```

```python
df2.to_excel("./data/bicycle_save.xlsx", index = False)
```

```python
df3.to_json("./data/read_save.json")
```
