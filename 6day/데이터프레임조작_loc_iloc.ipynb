```python
# 데이터 통합(행/열)
s1 = pd.Series([0,1],index=["A","b"])
s2 = pd.Series([2,3,4],index=["a","b","c"])

# 행단위 통합 : 기존 동일
# axis = 0 생략가능
pd.concat([s1,s2], axis=0)
```

```python
df_temp1 = pd.DataFrame([["Dog",3],["Cat",10],["Tiger",6],["Mouse",3]],
                        index=[0,1,2,3],
                        columns=["동물","나이"])
df_temp1
```

```python
df_temp2 = pd.DataFrame([["집",0],["초원",1],["수풀",0],["초원",1]],
                        index=[0,1,2,3],
                        columns=["사는곳","뿔개수"])
df_temp2
```

```python
# 열단위 통합 / 조건 없다 / 단, 행의 개수가 같은 경우 사용
pd.concat([df_temp1,df_temp2], axis=1)
```

```python
# 데이터프레임 준비
df_info1 = pd.DataFrame({'고객번호' : [1001,1002,1003,1004,1005,1006,1007],
                         '이름' : ['강','김','홍','이','박','최','소']},
                        columns = ['고객번호','이름'])
df_info1
```

```python
df_info2 = pd.DataFrame({'고객번호' : [1001,1001,1005,1006,1008,1010],
                         '금액' : [10000,20000,15000,5000,100000,30000]},
                        columns = ['고객번호','금액'])
df_info2
```

```python
# 데이터 옆으로 통합하기(조건)
# 두개 데이터프레임에서 같은 조건에 대해서만 가져오기
pd.merge(df_info1,df_info2,how="inner")
```

```python
# 무조건 갖고오기 / 왼쪽을 기준으로 모든 데이터 가져오고, 없으면 null
pd.merge(df_info1,df_info2,how="outer")
```

```python
pd.merge(df_info1,df_info2,how="left")
```

```python
pd.merge(df_info1,df_info2,how="right")
```
