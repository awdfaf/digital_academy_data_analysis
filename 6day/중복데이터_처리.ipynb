```python
# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
# 파일 불러오기
# encoding : 수집데이터가 utf-8 포멧이 아닌경우 오류발생 / 한글 데이터가 주로 오류 발생 -> euc-kr 설정
file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path,encoding="euc-kr")
df
```

```python
# 전체 컬럼들이 다음행에서 중복이 있는지 확인
df[df.duplicated()]
```

```python
# 특정 컬럼단위로 중복 체크
df[df.이용거리.duplicated()]
```

```python
# 두개 컬럼 중복 체크
df.duplicated(["이용거리","나이"])
```

```python
# 중복데이터 처리 방법 지정 : keep 속성 사용
# first : 첫번째 중복데이터는 살리고(false), 아래나머지 true
# last : 마지막 중복데이터는 살리고(false), 위에 나머지 true
df.duplicated(["이용거리"],keep="first")
```

```python
# 중복데이터 삭제
df.drop_duplicates(["이용거리"],keep="first")
```

```python
# 중복이 있는 모든 데이터 삭제하기
df.drop_duplicates(["이용거리"],keep=False)
```
