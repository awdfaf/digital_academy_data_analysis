```python
# 데이터 통합
# 컬럼단위 통합(옆으로, 조건에 맞게, 다른 파일을)
# 나라이름 가져오기 

# 파일 불러오기
code_master = pd.read_excel("./files_01/sample_codemaster.xlsx")
code_master
```

```python
sample_1
```

```python
# 컬럼단위로 통합
# left : 통합할 기준 데이터프레임
# right : 가져올 데이터가 있는 데이터프레임
# how : 어떻게 가져올지 지정(inner : 같은것만, left : 왼쪽전체, 없으면 없는대로)
# left_on : 왼쪽 데이터프레임에서 오른쪽 데이터프레임과 비교할 컬럼 지정
# right_on : 오른쪽 데이터프레임에서 왼쪽 데이터프레임과 비교할 컬럼 지정
# A31는 손실됨
pd.merge(left = sample_1,
         right = code_master,
         how = "inner",
         left_on = "국적코드",
         right_on = "국적코드")
```

```python
# how : left / A31 존재
pd.merge(left = sample_1,
         right = code_master,
         how = "left",
         left_on = "국적코드",
         right_on = "국적코드")
```

```python
# 변수생성
sample_1_code = pd.merge(left = sample_1,
                         right = code_master,
                         how = "left",
                         left_on = "국적코드",
                         right_on = "국적코드")
sample_1_code
```

```python
# 행단위 통합

# 데이터 불러오기
sample_2 = pd.read_excel("./files_01/sample_2.xlsx", header = 1, skipfooter = 2, usecols = "A:C")
sample_2
```

```python
# 문제
# sample_1의 최종결과와 동일한 컬럼을 만들어라
# 기준년월은 2019-12
# sample_2_code 변수로 국가이름 추가

sample_2["기준년월"] = "2019-12"

sample_2_code = pd.merge(left = sample_2,
                         right = code_master,
                         how = "left",
                         left_on = "국적코드",
                         right_on = "국적코드")
sample_2_code
```

```python
# 행단위 통합
#pd.concat([sample_1_code, sample_2_code])
sample = pd.concat([sample_1_code, sample_2_code], ignore_index=True)
sample
```

```python
# 파일 저장

sample.to_excel("./files_01_new/sample.xlsx")
```

```python
# 인덱스 지우기

sample.to_excel("./files_01_new/sample.xlsx", index = False)
```

```python
# files_all의 모든 데이터 통합하기
# 최종 결과 변수명 : sample_all
# 최종 결과 파일명 : sample_all.xlsx
# 파일 저장위치 : files_01_new 폴더
# 2019-01 부터
def integration():
    y = 19
    m = 1
    cnt = 0 
    # 국적 파일 불러오기
    code_master = pd.read_excel("./files_01/sample_codemaster.xlsx")
    
    # 첫번째 파일 불러오기
    file_first_name = "./files_01_all/sample({}).xlsx".format(0)
    file_first = pd.read_excel(file_first_name, header=1, skipfooter=2, usecols ="A:C")
    # 파일에 기준년월 컬럼 추가
    file_first["기준년월"] = "20{}-{}".format(str(y).zfill(2), str(m).zfill(2))
    # 파일에 국적코드 추가
    file_first = pd.merge(left =  file_first,
                          right = code_master,
                          how = "left",
                          left_on = "국적코드",
                          right_on = "국적코드")
    
    for num in range(37):
        # 다음 파일 불러오기
        # list.append로 할 수 있다
        file_next_name = "./files_01_all/sample({}).xlsx".format(num+1)
        file_next = pd.read_excel(file_next_name, header=1, skipfooter=2, usecols ="A:C")
        m = m + 1
        if m == 13:
            m = 1
            y += 1
        # 파일에 기준년월 컬럼 추가
        file_next["기준년월"] = "20{}-{}".format(str(y).zfill(2), str(m).zfill(2))
        # 파일에 국적코드 추가
        file_next = pd.merge(left =  file_next,
                              right = code_master,
                              how = "left",
                              left_on = "국적코드",
                              right_on = "국적코드")
        file_first= pd.concat([file_first, file_next], ignore_index=True)
        
       
        
    sample_all = file_first
    sample_all.to_excel("./files_01_new/sample_all.xlsx", index = False)
    return sample_all
    
    
    
```
