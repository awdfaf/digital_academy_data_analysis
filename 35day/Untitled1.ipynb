```python
# 파일 불러오기
# 데이터프레임 변수 : raw_total
import pandas as pd

file_path = "./files/3_1_crawling_raw.xlsx"
raw_total = pd.read_excel(file_path)
raw_total
```

```python
# tags 데이터 추출
raw_total["tags"]
```

```python
raw_total["tags"][0].find("#")
```

```python
raw_total["tags"][0].find(",")
```

```python
tag_total = []
for tags in raw_total["tags"]:
    split_list = tags[2:-2].split("', '")
    for tag in split_list:
        tag_total.append(tag)
```

```python
tag_total
```

```python
# 형태소 분석
from konlpy.tag import Okt
# 단어별 개수 카운트
from collections import Counter
# 단어빈도분석 : 워드클라우드 그리기
from wordcloud import WordCloud

# 워드 카운트
tag_counts = Counter(tag_total)
```

```python
tag_counts
```

```python
# 카운트 수가 가장 많은 상위 50개 조회
tag_counts.most_common(50)
```

```python
# 정제하기 : 불용어 처리
# 불용어를 제외시키기 위한 리스트 만들기
STOPWORD = ["#일상", "#선팔", "#제주도", "#jeju", "#반영구", "#제주자연눈썹",
            "#서귀포눈썹문신", "#제주눈썹문신", "#소통", "#맞팔", ""]
```

```python
# tags_total 워드카운트 데이터에서 불용어 제외
tags_total_selected = []
for tag in tag_total:
    if tag not in STOPWORD:
        tags_total_selected.append(tag)
tags_total_selected
```

```python
# 불용어 처리 후 워크카운트 실행
tags_total_selected = Counter(tags_total_selected)
tags_total_selected
```

```python
# 시각화 하기
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import platform

# 폰트설정
path = "c:/Windows/Fonts/malgun.ttf"
# 폰트지정
font_name = font_manager.FontProperties(fname=path).get_name()
rc("font", family = font_name)
```

```python
# 워드카운트 상위 30개 조회
tags_total_selected.most_common(30)
```

```python
# 워드카운트 상위 30개를 데이터프레임으로 만들기
# 워드카운터에는 리스트에 튜플형태로 되어있으며 각 튜플에는 2개의 데이터가 들어있음
tag_counts_df = pd.DataFrame(tags_total_selected.most_common(30))

# 컬럼명 변경하기 
tag_counts_df.columns = ["tags", "counts"]
tag_counts_df
```

```python
# 막대그래프 그리기
plt.figure(figsize=(10,8))
sns.barplot(x="counts", y="tags", data=tag_counts_df)
```

```python
# 워드클라우드 그리기
from wordcloud import WordCloud
font_path = "c:/Windows/fonts/malgun.ttf"
```

```python
# 배경색 : 흰색, 단어수 : 10개, 너비/높이 : 600/600, 그래프 사이즈 : 20, 15
# 이미지 저장 파일명 : "./img/jeju_tag_wordcloud.png"

wordcloud = WordCloud(
            font_path=font_path,
            background_color="white",
            max_words=10,
            relative_scaling=0.1, # 빈도 갯수에 따라 
            width=600,
            height=600
).generate_from_frequencies(tags_total_selected)

plt.figure(figsize=(20,15))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("./img/jeju_tag_wordcloud.png")
```
