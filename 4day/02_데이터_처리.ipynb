```python
def integration():
    # 초기 변수 생성
    df = pd.DataFrame()
    lst = []
    y = 2010
    m = 0
    n = 1000
    try:
        for i in range(n):
            # 파일 불러오기
            m += 1
            if m == 13:
                m = 1
                y += 1
            file_next_name = "./files_02/kto_{}{}.xlsx".format(str(y).zfill(2), str(m).zfill(2))
            file_next = pd.read_excel(file_next_name, header=1, skipfooter=4, usecols="A:G")
            # 기준년월 컬럼 추가
            file_next["기준년월"]="{}-{}".format(str(y).zfill(2), str(m).zfill(2))
            # 대륙 데이터 삭제
            filterlst = ["아시아주", "미주", "구주", "대양주", "아프리카주", "기타대륙", "교포소계"]
            condition = file_next["국적"].isin(filterlst) == False
            file_next = file_next[condition]
            #print(len(file_first)) 60개 확인
            #인덱스 재배열
            file_next = file_next.reset_index(drop=True)
            #대륙 컬럼 생성
            a = ["아시아"]*25 + ["아메리카"]*5 + ["유럽"]*23 + ["오세아니아"]*3 + ["아프리카"]*2 + ["기타대륙"] + ["교포"]
            file_next["대륙"] = a
            #관광 비율 컬럼 추가
            ssratio = round(((file_first["관광"] / file_first["계"]) * 100),0)
            file_next["관광객비율(%)"] = ssratio
            # 전체 국가 대비 관광객 비율 컬럼 추가
            allssratio = round(((file_next["관광"] / file_next["관광"].sum()) * 100),4)
            # allssratio.sum() 검증 : 합하면 1 / 반올림했기때문에 1을 넘을 수 있다.
            file_next["전체 비율(%)"] = allssratio
            # 리스트에 데이터프레임 삽입
            lst.append(file_next)
    except:
        pass
    
    try:
        # 리스트안에 있는 데이터프레임 통합
        first_file = lst[0]
        for i in range(n):
            first_file = pd.concat([first_file, lst[i+1]], ignore_index=True)
    except:
        pass
    df = first_file
    # 저장
    df.to_excel("./kto_total.xlsx", index = False)
    
    return df
```
